## `.apply`
- apply parameters directly without the need to create `ImageView` each time.

```python
class MainActivity : AppCompatActivity() {
    override fun onCreate(savedInstanceState: Bundle?) {
        super.onCreate(savedInstanceState)
        setContentView(R.layout.activity_main)

        val iv_image : ImageView = findViewById(R.id.iv_image)
        iv_image.setOnClickListener {
            iv_image.animate().apply {
                duration = 1000
                rotationYBy(360f)
            }.withEndAction{  # finish animation with another anim.

                iv_image.animate().apply {
                    duration = 1000
                    rotationYBy(1800f)
                }.start()  # starts the whole clicklistner
            }
        }
    }
}
```
