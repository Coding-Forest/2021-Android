## `View` and `Behavior` objects
We need to objects to visualise and functionalise (in other words, to create a working) `BottomSheet`. 
- 1) `View bottomSheet`
- 2) `BottomSheetBehavior BSBehavior`

```python
private BottomSheetBehavior BSBehavior;
View bottomSheet = findViewById(R.id.bottom_sheet);

BSBehavior = BottomSheetBehavior.from(bottomSheet);
```

## `.setBottomSheetCallback()`
- Sets a callback to be notified of bottom sheet events.

## .setState()

### `STATE_`
  - `BSB.STATE_DRAGGING`
  - `BSB.STATE_SETTLING`
  - `BSB.STATE_EXPANDED`
  - `BSB.STATE_COLLAPSED`
  - `BSB.STATE_HIDDEN`

```python
BSBehavior.setBottomSheetCallback(new BottomSheetBehavior.BottomSheetCallback() {
            @Override
            public void onStateChanged(@NonNull View bottomSheet, int newState) {
                switch (newState) {
                    case BottomSheetBehavior.STATE_DRAGGING:
                        TVState.setText("Dragging");
                        break;
                        
                        ...
                        
                    case BottomSheetBehavior.STATE_HIDDEN:
                        TVState.setText("Hidden");
                        break;
                }

            }

            @Override
            public void onSlide(@NonNull View bottomSheet, float slideOffset) {
                TVState.setText("Sliding~");
            }
        });
```

### `onStateChanged` vs `onSlide`
