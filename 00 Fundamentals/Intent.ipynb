## `intent.putExtra()`
- `putExtra()` is used when you want to send an extra data along with the `intent`.

- Reference: https://blog.naver.com/twotoedsloth/220012773872

```python
# send intent and some data along.
Intent intent = new Intent(Intent.ACTION_SEND);
intent.putExtra(Intent.EXTRA_EMAIL, recipients);
intent.putExtra(Intent.EXTRA_SUBJECT, subject);
intent.putExtra(Intent.EXTRA_TEXT, message);
```

```python
# receive intent and get extra data.
Intent intent = getIntent();
String[] recipients = intent.getExtras().getString("recipients");
String message = intent.getExtras().getString("subject");
String message = intent.getExtras().getString("message");
```
